# Task 1: Get Info Box and Store in a Python Dictionary

### Import Necessary Libraries

```python
from bs4 import BeautifulSoup as bs
import re
import requests
import pandas as pd
```

### Load our first page

```python
#Load the webpage
r = requests.get("https://pt.wikipedia.org/wiki/Big_Brother_Brasil_1")

#BeautifulSoup convertion
soup = bs(r.content)

#Print our HTML
print(soup.prettify())
```

### Search for the Names

```python
info_table = soup.find(class_="wikitable")
info_rows = info_table.find_all("tr")
for row in info_rows:
    print(row.prettify())
```

```python
participants_name = []

# Iterar sobre as linhas de info_rows e adicionar os nomes à lista
for row in info_rows[1:]:
    name_text = row.get_text().strip().split('\n')[0]
    participants_name.append(name_text)

# Exibir a lista resultante
participants_name
```

### Search for the Date of Birth

```python
participants_date = []

# Iterar sobre as linhas de info_rows e adicionar as datas à lista
for row in info_rows[1:]:
    date_text = row.get_text().split('\n\n')[1]
    participants_date.append(date_text)

# Exibir a lista resultante
participants_date
```

### Search for the Ocupation

```python
participants_ocupation = []

# Iterar sobre as linhas de info_rows e adicionar as ocupações à lista
for row in info_rows[1:]:
    occupation_text = row.get_text().split('\n\n')[2]
    participants_ocupation.append(occupation_text)

# Exibir a lista resultante
participants_ocupation
```

### Search for the Origin

```python
participants_origin = []

# Iterar sobre as linhas de info_rows e adicionar as origens à lista
for row in info_rows[1:]:
    origin_text = row.get_text().split('\n\n')[3]
    participants_origin.append(origin_text)

# Exibir a lista resultante
participants_origin
```

### Search for the Result

```python
participants_result = []

# Iterar sobre as linhas de info_rows e adicionar os resultados à lista
for row in info_rows[1:]:
    result_text = row.get_text().split('\n\n')[4].replace('em', ' em').replace('Vencedor', '1 lugar')
    participants_result.append(result_text)

# Exibir a lista resultante
participants_result
```

### Getting the Number of Votes

```python
info_table_second = soup.find_all(class_="wikitable")

if len(info_table_second) >= 2:
    second_wikitable = info_table_second[1]

    second_info_row = second_wikitable.find_all("td", {"bgcolor": True})
    print(second_info_row)
```

```python
votes = []

for row in second_info_row:
    text = row.get_text()  # Obtém o texto do elemento HTML
    numbers = [int(char) for char in text.split() if char.isdigit() and 1 <= int(char) <= 99]
    votes.extend(numbers)

print(votes)
```

### Getting the Astral Sign

```python
# Função para calcular o signo com base na data de nascimento
def signo(data_nascimento):
    dia, mes, _ = data_nascimento.split("/")
    
    if (mes == "03" and int(dia) >= 21) or (mes == "04" and int(dia) <= 19):
        return "Áries"
    elif (mes == "04" and int(dia) >= 20) or (mes == "05" and int(dia) <= 20):
        return "Touro"
    elif (mes == "05" and int(dia) >= 21) or (mes == "06" and int(dia) <= 20):
        return "Gêmeos"
    elif (mes == "06" and int(dia) >= 21) or (mes == "07" and int(dia) <= 22):
        return "Câncer"
    elif (mes == "07" and int(dia) >= 23) or (mes == "08" and int(dia) <= 22):
        return "Leão"
    elif (mes == "08" and int(dia) >= 23) or (mes == "09" and int(dia) <= 22):
        return "Virgem"
    elif (mes == "09" and int(dia) >= 23) or (mes == "10" and int(dia) <= 22):
        return "Libra"
    elif (mes == "10" and int(dia) >= 23) or (mes == "11" and int(dia) <= 21):
        return "Escorpião"
    elif (mes == "11" and int(dia) >= 22) or (mes == "12" and int(dia) <= 21):
        return "Sagitário"
    elif (mes == "12" and int(dia) >= 22) or (mes == "01" and int(dia) <= 19):
        return "Capricórnio"
    elif (mes == "01" and int(dia) >= 20) or (mes == "02" and int(dia) <= 18):
        return "Aquário"
    elif (mes == "02" and int(dia) >= 19) or (mes == "03" and int(dia) <= 20):
        return "Peixes"
    else:
        return "Data de nascimento inválida"

participants_sign = []  # Lista vazia para armazenar os signos das pessoas

# Calcular e adicionar o signo para cada participante
for data_nascimento in participants_date:
    signo_atual = signo(data_nascimento)
    participants_sign.append(signo_atual)

# Exibir a lista de signos
print(participants_sign)
```

### Creating the Data Frame

```python
import pandas as pd

# Dados dos participantes
data = {
    'Nome': participants_name,
    'Data de Nascimento': participants_date,
    'Signo': participants_sign,
    'Ocupação': participants_ocupation,
    'Origem': participants_origin,
    'Resultado': participants_result,
    'Número de Votos': votes
}

# Criar o DataFrame
df = pd.DataFrame(data)

# Mostrar o DataFrame
df
```
